```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/GeneralEsportData.csv")
```

#1. Visualisasi Data yang Relevan dengan Business Understanding

Tujuan studi kasus: Menganalisis peningkatan Genre dan Pendapatan pada tahun rilis

#Top 15 Genre game 1998-2020

```python
top_res = df['Genre'].value_counts().nlargest(15).index
plt.figure(figsize=(10,6))
sns.countplot(data=df[df['Genre'].isin(top_res)], y='Genre', order=top_res)
plt.title('Top 15 Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

#Tahun rilis serta pendapatan dari game terserbut

```python
plt.figure(figsize=(10,6))
sns.kdeplot(df[df['TotalEarnings'] < 500]['ReleaseDate'], bw_adjust=0.5)
plt.title('Distribution of Total Earnings')
plt.xlabel('Total Earnings')
plt.ylabel('Density')
plt.show()
```
